```python
def multi_graf(rows,cols):
    fig, ax = plt.subplots(nrows=rows, ncols=cols)
    for i in range(cols):
        for j in range(rows):
            ax[i,j].plot(age,salary)
            plt.tight_layout()
multi_graf(2,2)
```

```python
day = tips.day.unique()
total_bill = tips.groupby("day")["total_bill"].sum()

fig, ax = plt.subplots()
ax.bar(day,total_bill)
plt.show()
```

```python
day = tips.groupby('day')['total_bill'].sum().index
total_bill = tips.groupby('day').sum()['total_bill']
plt.bar(day,total_bill)
```

```python
day = tips.groupby("day")["total_bill"].sum().index
tip = tips.groupby("day")["tip"].sum()
fig, ax = plt.subplots()
ax.bar(day,tip)
plt.show()
```

```python
sex = tips.groupby("sex")["total_bill"].sum().index
tip = tips.groupby("sex")["tip"].sum()
fig, ax = plt.subplots()
ax.bar(sex,tip)
plt.show()
```

```python
day = tips.groupby("sex")["total_bill"].sum().index
tip = tips.groupby("sex")["total_bill"].sum()
fig, ax = plt.subplots()
ax.bar(sex,tip)
plt.show()
```

```python
plt.bar(tips.groupby('sex').total_bill.sum().index,tips.groupby('sex').total_bill.sum())
```

```python
day = tips.groupby("day")["total_bill"].sum().index
tip = tips.groupby("day")["tip"].sum()
total = tips.groupby("day")["total_bill"].sum()
fig, ax = plt.subplots()
ax.bar(day, total, label="total" )
ax.bar(day, tip, label="tip" )
plt.legend()
plt.show()
```

```python
day = tips.groupby("day")["total_bill"].sum().index
tip = tips.groupby("day")["tip"].sum()
total = tips.groupby("day")["total_bill"].sum()

fig, ax = plt.subplots()
ax.bar(day, tip, label="tip", align="edge")
ax.bar(day, total, label="total",width=0.6, align="edge")

plt.legend()
plt.show()
```
