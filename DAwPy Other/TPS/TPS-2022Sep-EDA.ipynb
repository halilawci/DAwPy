## Year ❌ Country ❌ Num Sold

<div style="padding:10px;color:black;margin:0;font-size:125%;display:fill;border-radius:5px;background-color:Gold;overflow:hidden;font-weight:700;border: 1px solid deepskyblue;">󠁧󠁢Year ❌ Country ❌ Num Sold</div>

```python
g = sns.catplot(row = "country", y = "num_sold", data = df, kind = "bar", x = "year", sharex=False)

for ax in g.axes.ravel() :
    for i in ax.containers :
        ax.bar_label(i)
plt.suptitle("Yıllara ve Ülkelere göre Num Sold Ortalamaları", size = 15)
g.fig.subplots_adjust(top = 0.95)
```

```python
ax = sns.barplot(y = df.num_sold, x = df.country, hue = df.year, ci = None)

for i in ax.containers :
    plt.bar_label(i, fmt = "%2.f")
plt.suptitle("Yıllara ve Ülkelere göre Num Sold Ortalamaları", size = 15)
plt.legend(loc = (1,0.76));
```

**Yıllara ve Ülkelere göre Num Sold Ortalamaları 2 şekilde de görüldüğü gibidir**

```python
df.groupby(["year", "country"]).num_sold.mean().unstack(level = 0).plot(kind = "pie",
                                                                        subplots = True,
                                                                        layout = (2,3),
                                                                        figsize = (20,12),
                                                                        legend = False,
                                                                        ylabel = "",
                                                                        title = df.groupby(["year", "country"]).num_sold.mean().unstack().index.tolist(),
                                                                        autopct = "%.2f%%",
                                                                        colors = ["Salmon", "Gold", "violet", "lightgreen", "CornflowerBlue", "Bisque"]);
```
