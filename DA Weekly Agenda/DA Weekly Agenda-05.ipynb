# 10. Set an additional dimension using size column

```python
# CODE HERE
```

```python
plt.scatter(x=tips.tip, y=tips.total_bill,c = tips["size"], cmap='coolwarm') ## sequence of colors 
z=plt.colorbar()# color scale bar 
z.set_label('size')##
```

```python
plt.scatter(tips["tip"], tips["total_bill"], linewidth = 0.001, c = tips["size"], cmap='coolwarm')
plt.colorbar()
plt.xlabel("tip")
plt.ylabel("total_bill")
plt.show()
```
